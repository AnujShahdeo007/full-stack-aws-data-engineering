# Print() 

Syntax: 
 
print(*object,sep='',end='\n',file=sys.stdout,flush=False)

```python
a=10
b=20
c=a+b
```

```python
print(c)
```

```python
print("hello") # sysout
```

```python
print("Spark","Python",sep='-')
print("new object")
```

```python
print("hello")
print("world")
```

```python
print("hello",end=" ")
print("World",end="\n")
```

```python
for i in range(5):
    print(i,end=",")
```

```python
f=open("output.txt","w")
```

```python
print("Hello",file=f)
f.close()
```

```python
with open("log.txt","a") as log:
    print("Job started",file=log)
```

```python
import time
for i in range(5):
    print(i,end=" ",flush=True)
    time.sleep(1)
```

```python
from datetime import datetime
print(
    datetime.now(),
    "INFO",
    "Job Started",
    sep="|"
)
```
