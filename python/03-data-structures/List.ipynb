# Shallow Copy 

- Shallow copy creates a new Outer Object but does not recursively copy the objects inside it.

- Instead, the new outer object contains refrence to the same inner objects as Original 

# The Outer Container is copied 
# The nested (Inner) Object are shared

```python
numbers=[1,2,3,2,5,1]
for i in range(len(numbers)):
    duplicate=False
    # Check if current number has already appered 
    for k in range(i):
        if numbers[i]==numbers[k]:
            duplicate=True
            break
    if duplicate:
        continue
    # Check if it is appers later 
    for j in range(i+1,len(numbers)):
        if numbers[i]==numbers[j]:
            print(numbers[i])
            break
```
