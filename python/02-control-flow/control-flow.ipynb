# Syntax for Match-case 

match Expression:
    case value 1:
        Statements 
    case value 2:
        statements
    case value 3:
        statements
    case_:
        default statement 
        

```python
day=int(input("Enter the day of the week (1-5): "))
match day:
    case 1:
        print("Monday")
    case 2:
        print("Tuesday")
    case 3:
        print("Wednesday")
    case 4:
        print("Thursday")
    case 5:
        print("Friday")
    case _:
        print("Weekend")
```

```python
operator = input("Enter an operator (+, -, *, /): ")
match operator:
    case "+":
        print("Addition")
    case "-":
        print("Subtraction")
    case "*":
        print("Multiplication")
    case "/":
        print("Division")
    case _:
        print("Invalid operator")
```

```python
Environment = input("Enter the environment (dev, test, prod): ")
match Environment:
    case "dev":
        print("Deploy to development environment")
    case "test":
        print("Deploy to test environment")
    case "prod":
        print("Deploy to production environment")
    case _:
        print("Invalid environment")
```

```python
services="Glue"
match services:
    case "S3":
        print("Storage Service")
    case "Glue":
        print("Data Integration Service")
    case _:
        print("Unknown Service")
```
